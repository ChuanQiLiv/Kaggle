```python
q25,q75 = np.percentile(df['Customer_Age'],[25,75])
iqr = q75 - q25
low = q25-(1.5*iqr)
high = q75+(1.5*iqr)
low,high
```

```python
df.loc[(df['Customer_Age'] >= low) & (df['Customer_Age'] <= high)]
```
